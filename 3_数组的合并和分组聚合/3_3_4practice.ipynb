# 1.使用matplotlib呈现出店铺总数排名前10的国家

```python
from matplotlib import pyplot as plt
from matplotlib import rc
import pandas as pd
import numpy as np
```

```python
file_path = './starbucks_store_worldwide.csv'
```

```python
df = pd.read_csv(file_path)
print(df.head(1))
```

```python
data1 = df.groupby(by='Country').count()['Brand'].sort_values(ascending=False)[:10]
print(data1)
```

```python
_x = data1.index
_y = data1.values
```

```python
font = {'family' : 'simsun',
          'weight' : 'bold'}
rc('font',**font)
```

```python
#画图
plt.figure(figsize=(20,10),dpi = 80)
plt.bar(range(len(_x)),_y)
plt.xticks(range(len(_x)),_x)
plt.grid(alpha = 0.3)
plt.show()
```

# 2.使用matplotlib呈现出中国每个城市的店铺数量

```python
df = df[df['Country'] == 'CN']
# print(df.head(1))
```

```python
data2 = df.groupby(by='City').count()['Brand'].sort_values(ascending=False)[:25]
print(data2)
```

```python
_x = data2.index
_y = data2.values
```

```python
#画图
plt.figure(figsize=(20,10),dpi = 180)

# plt.bar(range(len(_x)),_y,width = 0.3)
plt.barh(range(len(_x)),_y,height = 0.3)
# plt.xticks(range(len(_x)),_x,rotation = 325)
plt.yticks(range(len(_x)),_x)
plt.grid(alpha = 0.3)
plt.show()
```
